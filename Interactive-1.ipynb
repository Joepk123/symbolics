```python
import sympy as sp
from symbolics import LinearDifferentialOperator, ExpandableFunction

# 1. Define our spatial coordinate
x = sp.Symbol('x')

# 2. Define a quick test wavefunction (using your base class!)
class MyTestWave(ExpandableFunction):
    def __new__(cls, coord, **kwargs):
        return super().__new__(cls, coord, **kwargs)
    
    @property
    def definition(self):
        # Let's use sin(x) as the explicit math
        return sp.sin(self.args[0])

# 3. Create the mathematical objects
wave = MyTestWave(x)

# Create an operator: L = d^2/dx^2 + 5
L = LinearDifferentialOperator(x, {
    2: 1,  # 1 * d^2/dx^2
    0: 5   # 5 * (no derivative)
})

# 4. Apply the operator to the wave
# This creates the abstract AST tree: L(MyTestWave(x))
equation = L(wave)

print("--- Step 1: Abstract Equation ---")
print(equation) 
# Should print something abstract, hiding the heavy math!

# 5. Execute the engine!
print("\n--- Step 2: Evaluated Result ---")
result = equation.evaluate()
print(result) 
# The engine should output: 4*sin(x)
# Because: d^2/dx^2 sin(x) + 5*sin(x) = -sin(x) + 5*sin(x) = 4*sin(x)
```
